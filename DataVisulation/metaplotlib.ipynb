```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
# matplotlib basic
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.plot(y,x)
```

```python
#               subplots in matplotlib
plt.subplot(2,2,1)
plt.plot(x,y)  
plt.subplot(2,2,2)
plt.plot(y,x)  
plt.subplot(2,2,3)
plt.plot(x**2,y)  
plt.subplot(2,2,4)
plt.plot(x,y**2)  
```

```python
#object oriented way
fig = plt.figure(figsize=(8,7) , dpi= 200)# from (dpi(clearly increase the quality))figsize you can increase the width and height 
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,.3,.3])
axis2.plot(x,y)
```
