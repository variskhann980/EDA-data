```python
#We use distribution plots to visualize the distribution of quantitative data like numeric data.

#Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

#distplot/histplot
#jointplot
#pairplo    t
#rugplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips') 
df
```

```python
df['size'].unique()
```

```python
#distplot/histplot
sns.histplot(df['total_bill'], bins = 5 , kde=True)
```

```python
#jointplot
sns.jointplot(data=df , x= "total_bill" , y="tip" , kind="reg" ,color="red")
```

```python
#pairplot
sns.pairplot(df,hue="sex", palette="rainbow")
```

```python
sns.rugplot(df["total_bill"])
```
