```python
import numpy as np
import pandas as pd 
#                   Finding Missing Data
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df.isna()

```

```python
missingData = df.isna().sum() # this is give the number of null data in your data 
missingData
```

```python
checking = df.isin([1 ,2 ,3 ])
checking
```

```python
df.isna().any()# any means NAN value is exist or not
```

```python
df.dropna()# it is show the that row where the Nan not exit 
```

```python
df.dropna(thresh=3) # it is remove those row value which is less than 3
```

```python
#                   Filling the missing Data
df.fillna(0)# it is replace the NAN and fill the 0
```

```python
values = {"A" : 0 , "B" : 100 , "C" : 200 , "D" : 500}
df.fillna(value=values)
```
